```python
import pandas as pd
import numpy as np
import csv
books = pd.read_csv("book_csv_data.csv",encoding='ISO-8859-1')
```

```python
books.head(5)
```

```python
books = books.drop_duplicates()
```

```python
books.head(4)['Description']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv=TfidfVectorizer(min_df=3, max_features = None,
                    strip_accents = 'unicode', analyzer = 'word', token_pattern=r'\w{1,}',
                    ngram_range = (1,3),
                    stop_words = 'english')

books['Description']=books['Description'].fillna('')
```

```python
tfv_matrix = tfv.fit_transform(books['Description'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
sig[0]
```

```python
indices = pd.Series(books.index, index = books['Original_Title']).drop_duplicates()
```

```python
def give_rec(title):
    title=title.lower()
    print(title)
    try:
        idx=indices[title]
    except:
        print("Book Not present")
        return ['Book Not Present.']
    else:
        sig_scores = list(enumerate(sig[idx]))
        try:
            sig_scores = sorted(sig_scores, key=lambda x: x[1] , reverse=True)
        except:
            sig_scores = sorted(sig_scores, key=lambda x: x[1].all() , reverse=True)
        sig_scores = (sig_scores[1:11])
        book_indices = [i[0] for i in sig_scores]
        string=books['Original_Title'].iloc[book_indices].tolist()
        list1=[]
        for stg in string:
            try:
                list1.append(stg.title())
            except:
                list1.append(stg)
        print(list1)       
        return list1
```
