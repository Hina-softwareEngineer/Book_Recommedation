```python
import xlrd 

loc=(r'C:\Users\HINA KHADIM\Documents\docs\book tags.xlsx')
  
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(0) 
sheet.cell_value(1, 0) 
  
ids=[]
for i in range(sheet.nrows): 
    #print(sheet.cell_value(i, 0)) 
    ids.append(sheet.cell_value(i, 0))
```

```python
print(ids[1])
```

```python
import requests
import json
from bs4 import BeautifulSoup
import bs4

j=1
ID=[]
name=[]
count=[]

for iD in range(3707,len(ids)):
    result=requests.get("https://www.goodreads.com/book/isbn/"+ids[iD]+"?key=wSPrZyYr3bCAA185EXHlOg")
    soup=bs4.BeautifulSoup(result.content,"html.parser")
    title=soup.find_all("shelf")
    print(iD)
    for i in range(0,len(title)):
        ID.append(iD)
        name.append(title[i].get("name"))
        count.append(title[i].get("count"))
    j=j+1
```

```python
import xlsxwriter
workbook   = xlsxwriter.Workbook(r'C:\Users\HINA KHADIM\Documents\docs\ISBNS.xlsx')
worksheet = workbook.add_worksheet()
row = 0
col = 0

for i in range(3707,len(ids)):
    worksheet.write(row, col, ID[i])
    worksheet.write(row, col + 1, name[i])
    worksheet.write(row, col + 2, count[i])
    row += 1
    
workbook.close()
```
